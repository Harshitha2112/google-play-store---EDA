<span style="font-size:20px">**Which app has maximum reviews**</span>

```python
"""data['Reviews'].max() → finds the maximum value

Then compares:

max_value == each row value
Returns True only for rows where Reviews = max
"""
data[data['Reviews'].max()==data['Reviews']]
```

```python
#we are only interested in app name
data[data['Reviews'].max()==data['Reviews']]['App']
```

<span style="font-size:20px">**Display top 5 apps having highest reviews**</span>

```python
#fetch the index values of those rows and assign it to variable
index=data['Reviews'].sort_values(ascending=False).head().index
```

```python
#pass that varible to iloc to get full rows of that index
data.iloc[index]
```

```python
data.iloc[index]['App']
```

```python
#simplest and fastest way of sorting
data.nlargest(5, 'Reviews')['App']
```
